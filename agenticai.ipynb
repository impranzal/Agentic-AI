## Agentic AI

- It is AI system where we not only generate text but also consider goals, reasons about next steps, check results, works until your outcome is achieved. 
- It uses LLM as brain, retrival memory, and tools used to perform specific tasks.


## How it works
- Input:
    User gives their goal
- Reason:
    The model interprets the goal
- Act:
    Model calls for tool or performs steps of operation
- Observe 
    The system reads the result
- Repeat:
    It keep on going until all the task are completed

This loop is what makes system agentic 


## Agentic AI vs LLM(Large Language Models)
- A plain LLM mainly predicts next token in text and return the result based on prompts and consider the context.
- An AI agent uses LLM + Extra tools, memory, planning and feedbacks to generate a response.
eg:
prompt: "Write summary"
It return text

AI agent:
prompt: "Reserach competitors, summarize findings, and draft the email"

- It breaks the prompt into steps, searches using tools, gather findings and then write the email.

# Task 1:
## Build a CLI for student record using dictionary:
### Features:
    1. add <name> <marks>
    2. show -> display
    3. average -> average
    4. topper -> display the topper
    5. exit -> exit the system


````python
students = {}  ## store students record

while True:
    cmd = input("Enter your command: ")  # add roshan 21
    parts = cmd.split()   #["add", "roshan", "21"]
    if(len(parts) == 0):
        continue
    if(parts[0]== "add"):
        name = parts[1]
        marks = int(parts[2])
        
        students[name] = marks
        print(f"{name} addded with marks :{marks}\n")
    
    elif(parts[0] == "show"):
        if students:
            for name, marks in students.items():
                print(name, ": ", marks)
        else:
            print("No student record to display")
    elif(parts[0]== "average"):
        if students:
            avg = sum(students.values())/len(students)
            print("Average marks: ", avg)
        else:
            print("No students record to display")
    elif(parts[0]== "topper"):
        if students:
            top = max(students, key=students.get)
            print("Topper: ", top)
        else:
            print("No student record")
    elif cmd== "exit":
        print("Exiting student record system")
        break
        
````

## Chat history analyzer 
- Store the user message in a list
- Features:
    - Count repeated words
    - show most used words
    - show last 3 messages
    - i/p: messages = ['msg a', "msg b", ....]

````python
from collections import Counter

messages = []

while True:
    cmd = input("Enter command: ")

    if cmd == "add":
        messages.append(input("Enter message: "))

    elif cmd == "last3":
        print(messages[-3:])

    elif cmd == "count":
        words = []
        for m in messages:
            words += m.split()

        freq = Counter(words)

        for w in freq:
            print(w, ":", freq[w])

    elif cmd == "most used":
        words = []
        for m in messages:
            words += m.split()

        if words:
            freq = Counter(words)
            print(max(freq, key=freq.get))
        else:
            print("No data")

    elif cmd == "exit":
        break
````

### Develop a platform of agentic AI for an interviewer Taking DSA interview

````python
import requests 
ollama_url = "http://localhost:11434/api/chat"

System_prompt = """
 You are a strict Data Structures and Algorithms interviewer.

Rules:
- Ask only ONE question at a time.
- Start with easy questions and gradually increase difficulty.
- Wait for the user's answer before continuing.

Evaluation Rules (VERY IMPORTANT):
- Always evaluate the user's answer carefully.
- Do NOT assume the user is correct.
- If the answer is WRONG:
    - Clearly say: "Incorrect."
    - Briefly explain why it is wrong.
    - Then give the correct answer.
- If the answer is CORRECT:
    - Say: "Correct."
    - Optionally provide a better or more optimal answer.

Behavior Rules:
- Never say "partially correct".
- Never skip evaluation.
- Keep responses short, clear, and professional.
- Stay focused only on DSA topics.

Flow:
1. Ask question
2. Wait for answer
3. Evaluate strictly
4. Ask next question

Continue until the user types "exit".

"""

# memory
messages = [
    {"role": "system", "content":System_prompt}
]

def chat_with_ai(messages):
    response = requests.post(ollama_url, json = {
        "model": "llama3.1:8b",
        "messages": messages,
        "stream": False
    })
    return response.json()["message"]["content"]

print("Bot Interview started(type 'exit' to quit)")
messages.append({"role": "user", "content": "start interview"})
ai_reply = chat_with_ai(messages)

print("AI: ", ai_reply)
## ask question 
messages.append({"role": "assistant", "content": ai_reply})

# main loop:
while True:
    user_input = input("\n You: ")
    if user_input.lower() == "exit":
        print("Interview Ended")
        break

    # Store user response 
    messages.append({"role": "user", "content": user_input})
    ai_reply = chat_with_ai(messages)
    print("AI", ai_reply)

    # store AI reply
    messages.append({"role": "assistant", "content": ai_reply})
````

### Large Language Model (LLM)

- It is a sophisticated function used to predice the next word in the given piece of text.
- I am going to .........
    - Multiple parameters()
    - Probability
    - High
- Instead of choosing one word to find possibility of all next possible words.
- To make the output natural it works on probability even if it looks like a deterministic mode.
- It gives different output for similar prompts.

Parameters (Weights);
- These are numeric values that determines the behaviour of our model.
- LLMs can have hundreds of billions of these parameters

- Modern LLMs are ChatGPT(OpenAI), Google Gemini, Anthropic(Claude)

### Input (Text, Voice, Image) -> LLMs -> Output (Questioning & reasoning, sentement analysis, image captioning, object recognization) 

## Working of LLMs


### Input Embedding: Converts the text in vectors.
### Position Encoding: It adds sequence of informarion in orders.
### Self-attention: Understanding relationship between words in context.
### Feed forward Layers: Helps to capture complex patterns.
### Decoder: Generate response step by step
### Parallel relationship reasoning and generates output.

### Training process: 

1. Pre-training: The initial phase where a model process enormous amount of text(from internet) to learn patterns.
2. Back-propaation: It is an algorithm designed to train tweak parameters. If the model makes an incorrect predictionsthen weights are adjusted to make the correct words more likely in the future.
3. Scale of computations: The training involves a mind-boggling numebrs of operations. If a person performed one billion addition and multiplication in each seconds, then it will still take over hundred millions years to compute the operation for LLM.
4. Reinforrcement learning with human feedback(RLHF): It is a secondary training type where human workers flag unhelpful prediction. So model works like a helpful assistant rather than just text completer.
   

### Tokens

- It is a basic unit of text that the model processes.
- It can represent various components of language including:
    - Words: apple, entertainment
    - Subwords: Un - happy
    - Prefix
    - Suffix
    - Punctuations
    - Special Tokens
- Tokenization is the process of breaking down text into smaller units.

- GPT-4 has context window of 8,192 tokens with some  versional having more than 30k tokens.

## Temperature: 

- Temperature in LLM is a hyperparameters that controls the randomness and creativity of the generated text.
- Typically ranges betweel 0.0 to 2.0
- 0.0 - 0.5: lower temperature -> deterministic, focused, factual
- 0.6 - 1.0+: Increases diversity and creativity, ideal for brainstorming or creating writing.

## Key aspects:

### How does it work?

- It shapes the probability distributionof next token. A low temperature makes the model confidently select top- ranked words. A high tmeperature also cosiders words with low probability.
- Low temperature: Produces predictable, consrevative, repetitive outputs. Best for MAth, or factual Q&A
- High temperature: Increases randomness resulting more diverse, innovative model but the probability of getting false response is high.

## Context window: 

- A context window is the span of text that a model can consider at a time while making predictions.
- It is the lookback or the amount of previous informationthat a model uses to make sense of current input.



- LLMs such as GPT and Geminis heavily rely on context windows for predicting the next token in sequence.
- The larger the context window the more information model can access to understand the meaning of text.

````python
import ollama

def analyze_news(news_text):
    prompt =  f"""
    Analyze the following news article.

    Task: 
    1. Summarize in one sentence
    2. Classify category[political, technology, sports, business]
    3. Detect sentiments[positive, negative, neutral]
    4. Create a short heading(headline)

    Output in JSON:
    {{
    "summary" : "",
    "category" : "",
    "sentiment" : "",
    "headline" : ""
    }}

    News : {news_text}
    
    """

    response = ollama.chat(
        model = "llama3.1:8b",
        messages = [{"role":"user", "content":prompt}]
    )

    return response['message']['content']

# test

news = """
President Donald Trump threatened Saturday to “hit and obliterate” Iranian power plants if Tehran does not reopen the Strait of Hormuz within 48 hours.

“If Iran doesn’t FULLY OPEN, WITHOUT THREAT, the Strait of Hormuz, within 48 HOURS from this exact point in time, the United States of America will hit and obliterate their various POWER PLANTS, STARTING WITH THE BIGGEST ONE FIRST!” Trump wrote in a Truth Social post.

The threat marks an escalation in rhetoric from Trump, who’s previously dangled the option of hitting Iranian infrastructure but cautioned it would impair the country’s ability to rebuild. And it’s a tacit acknowledgement that the strait’s closure provides Iran significant leverage.

Trump’s latest threat comes a day after he said the US is considering “winding down” military efforts in the Middle East and dismissed questions about the White House’s plan for restoring traffic through the Strait of Hormuz, saying that “at a certain point, it’ll open itself.”

It also comes after Trump said Thursday that he’d told Israeli Prime Minister Benjamin Netanyahu to stop attacks on Iranian oil and gas facilities.
"""

result = analyze_news(news)
print(result)
````
