# Retrieval Augmented Generation (RAG)

- LLM models are feed with data and gives responses based on memorized data but it does not work with external files like csv, docs, etc. Now we need RAG.
- RAG pulls information from csv and then generate answer based on retrieved context.

- RAG is an advance framework of AI that combines information retrieved through text generation model like GPTs and Geminis to produce more accurate and relevant up-to-date response.


## Myths about RAG

- RAG is dead.
- Bigger the context window means you don't need RAG anymore.
- System (Millions of tokens).
    - cost is high.
    - latency increased and system becomes slow.
    - LLM performance is worst when it is overloaded.


# R -> Retrieve: Find useful information
# A -> Augment: Add it to AI knowledge
# G -> Generate: Create better response

## Components of RAG: 






### 1. Ingestion:

- Before retrieval documents must be processed and stored in a system
- THis process is requires breaking document into smaller pieces known as **chunks**
- A *native approach* is using fixed sized chunk. Fixed size chunks means splitting the document into equal token lenth(500 tokens per chunk)
- Main problems:
    - Loss of context at chunk boundry
    - If a sentene is splitted between two chunks then it becomes incomplete.
- So we use **Semantic Chunking**:
    - Here embedding model can be used to analyze the text.
    - Frameworks like **longChain and Llama Index** are built in to follow semantic chunking.
